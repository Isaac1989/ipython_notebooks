The clue shows the bull picture with the clue:

        len(a[30]) = ?
        
Looking at the source indicates there is another polygon embedded in the image. Clicking on it, forwards to 

http://www.pythonchallenge.com/pc/return/sequence.txt

(could have also got this from the source). The page displays:

        a = [1, 11, 21, 1211, 111221,
        
This is an example of a look-say algorithm, where you describe the previous entry:

        1 --> 'one 1' (11)
       11 --> 'two 1s' (21)
       21 --> 'one 2, one 1' (1211)
     1211 --> 'one 1, one 2, two 1s' (111211)
     
 This can be implemented with the following function:


```python
def looksay(look):
    prev, count, say = look[0], 1, ''
    for char in look[1:]:
        if char == prev:
            count += 1
            continue
        say += str(count) + prev
        prev = char
        count = 1
    return say + str(count) + prev

# NOW RUN RECURSIVELY 
look = '1'
for ii in range(30):
    look = looksay(look)

# len(a[30]) = ?
print len(look)
```
