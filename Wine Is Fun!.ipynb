```python
features_names = ['Alcohol', 'Malic acid', 'Ash', 'Alcalinity of ash', 'Magnesium', 'Total phenols', 
                'Flavanoids', 'Nonflavanoid phenols', 'Proanthocyanins', 'Color intensity', 
                'Hue', 'OD', 'Proline']
target_name = 'Region'
column_names = [ target_name ] + features_names
```

```python
data = pd.read_csv('wine.data', header=None, names=column_names)
data.head()
```

#Checking out the data

```python
from pandas.tools.plotting import scatter_matrix
```

```python
scatter_matrix(data);
```

```python
from pandas.tools.plotting import radviz
```

```python
radviz(data,'Region');
```

#Train Linear Support Vector Machine

```python
from sklearn.svm import LinearSVC

svc_model = LinearSVC(C=1.0).fit(data[features_names].values, data[target_name].values)
svc_predictions = svc_model.predict(data[features_names].values)
1.0 * sum(svc_predictions == data[target_name].values) / len(data[target_name].values)
```

#Using features scaled to have unit norm

```python
from sklearn import preprocessing
data_normalized = preprocessing.normalize(data[features_names].values)
```

```python
data_normalized = pd.DataFrame(data_normalized,columns=features_names)
```

```python
scatter_matrix(data_normalized);
```

#Train SVM on Normalized features
need to boost the regularization constant for these features

```python
svc_model = LinearSVC(C=300,loss='l2').fit(data_normalized[features_names].values, data[target_name].values)
svc_predictions = svc_model.predict(data_normalized[features_names].values)
1.0 * sum(svc_predictions == data[target_name].values) / len(data[target_name].values)
```
